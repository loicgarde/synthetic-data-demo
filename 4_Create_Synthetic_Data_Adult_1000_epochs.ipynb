***Rechargement du synthetiser préalablement entrainé***




```python
from sdv.single_table import CTGANSynthesizer

synthesizer = CTGANSynthesizer.load(
    filepath='./drive/MyDrive/Demo_Synthetic_Data/adult_synthesizer_1000epochs.pkl'
)
```

***Génération d'un jeu de données synthétiques (100 000 enregistrements)***

```python
synthetic_data = synthesizer.sample(
    num_rows=100000,
)
```

***Affichage du jeu de données réel***

```python
from sdv.datasets.demo import download_demo

real_data, metadata = download_demo(
    modality='single_table',
    dataset_name='adult'
)

real_data
```

***Affichage du jeu de données synthétiques***

```python
synthetic_data
```

***Evaluation de la qualités des données générées***

```python
from sdv.evaluation.single_table import evaluate_quality

quality_report = evaluate_quality(
    real_data,
    synthetic_data,
    metadata
)
```

***Evaluation de la qualité par colonne***

```python
quality_report.get_visualization('Column Shapes')
```

***Dispersion de données réelles versus synthétiques colonne 'education-num'***


```python
from sdv.evaluation.single_table import get_column_plot

fig = get_column_plot(
    real_data=real_data,
    synthetic_data=synthetic_data,
    column_name='education-num',
    metadata=metadata
)

fig.show()
```

***Corrélation colonnes 'education-num' et 'marital-status' (réelles versus synthétiques)***

```python
from sdv.evaluation.single_table import get_column_pair_plot

fig = get_column_pair_plot(
    real_data=real_data,
    synthetic_data=synthetic_data,
    column_names=['marital-status', 'education-num'],
    metadata=metadata
)

fig.show()
```
